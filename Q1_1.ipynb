**Simulate a Nonlinear Function using tensorflow.keras**


```
import numpy as np
import pandas as pd
import tensorflow as tf 
from tensorflow import keras
from matplotlib import pyplot as plt
from google.colab import files
```

# **Define Three nonlinear function**

```
def poly(x):
  return x**2+2*x+1 # polynomial function
def trig(x):
  return np.cos(x) # Trig function
def expo(x):
  return np.exp(-x) # exponential  function
```

```
# training data
x_train1=np.linspace(-7,5,200)

x_train2=np.arange(-2*np.pi,2*np.pi,np.pi/30)

x_train3=np.linspace(0,10,200)

# plot the function
```

```
y_train1=poly(x_train1)
y_train2=trig(x_train2)
y_train3=expo(x_train3)
```



In this question we 

```

fig =plt.figure(figsize=(20,8))
fig.suptitle('Polynomial, trignometric, and exponential Functions Considered',fontsize=30)
#ax1
ax1 = fig.add_subplot(1,3,1)
ax1.scatter(x_train1,y_train1,s=10,color='r', label='polynomial')
ax1.set_xlabel('X values',fontsize=16)
ax1.legend(loc='upper left')
ax1.set_ylabel('Polynomial function values',fontsize=16)
# ax2
ax2 = fig.add_subplot(1,3,2)
ax2.scatter(x_train2,y_train2,s=10, color='orange', label='Trignometric')
ax2.set_xlabel('X values',fontsize=16)
ax2.legend(loc='upper left')
ax2.set_ylabel('Trignometric function values',fontsize=16)
#ax3
ax3 = fig.add_subplot(1,3,3)
ax3.scatter(x_train3,y_train3,s=10,color='g', label='exponential')
ax3.set_xlabel('X values',fontsize=16)
ax3.set_ylabel('exponential function values',fontsize=16)
ax3.legend(loc='upper left')
```



# **Model #1 Shallow neural network**


*   1 hidden layer with 812 hidden units
*   Activation = ReLU
*   Optimizer = Adam
*   Mean Squared Error (MSE)

Number of epochs = 500


```
# for polynomial function
model_shallow1=tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(1,)),
                                  tf.keras.layers.Dense(870,activation=tf.nn.relu),
                                  tf.keras.layers.Dropout(0.5),
                                  tf.keras.layers.Dense(1)])
opt=tf.keras.optimizers.Adam(learning_rate=0.02)
model_shallow1.compile(optimizer=opt,loss='mse',metrics=['accuracy'])
hs1=model_shallow1.fit(x_train1,y_train1,epochs=100,batch_size=32,validation_split=0.3)  
model_shallow1.summary()
```

```
# for trignometric function
model_shallow2=tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(1,)),
                                  tf.keras.layers.Dense(870,activation=tf.nn.relu),
                                  tf.keras.layers.Dropout(0.5),
                                  tf.keras.layers.Dense(1)])
opt=tf.keras.optimizers.Adam(learning_rate=0.005)
model_shallow2.compile(optimizer=opt,loss='mse',metrics=['accuracy'])

hs2=model_shallow2.fit(x_train2,y_train2,epochs=100,batch_size=16,validation_split=0.3)
```

```
# for exponential function
model_shallow3=tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(1,)),
                                  tf.keras.layers.Dense(870,activation=tf.nn.relu),
                                  tf.keras.layers.Dropout(0.5),
                                  tf.keras.layers.Dense(1)])
opt=tf.keras.optimizers.Adam(learning_rate=0.02)
model_shallow3.compile(optimizer=opt,loss='mse',metrics=['accuracy'])
hs3=model_shallow3.fit(x_train3,y_train3,epochs=100,batch_size=64,validation_split=0.3)  
```

# **Model #2 Deep neural network (using 3 hidden layers)**






```
model_deep1=tf.keras.models.Sequential([#tf.keras.layers.Flatten(x),
                                  tf.keras.layers.Dense(128,activation=tf.nn.relu), # Hidden 1
                                  tf.keras.layers.Dropout(0.5),
                                  tf.keras.layers.Dense(16,activation=tf.nn.relu),# Hidden 2
                                  tf.keras.layers.Dense(16,activation=tf.nn.relu), # Hidden 3
                                  tf.keras.layers.Dense(1)]) # Output
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
model_deep1.compile(optimizer=opt,loss='mse',metrics=['accuracy'])
hd1=model_deep1.fit(x_train1,y_train1,epochs=100,batch_size=32)
```

```
model_deep2=tf.keras.models.Sequential([#tf.keras.layers.Flatten(x),
                                  tf.keras.layers.Dense(128,activation=tf.nn.relu), # Hidden 1
                                  tf.keras.layers.Dropout(0.5),
                                  tf.keras.layers.Dense(16,activation=tf.nn.relu),# Hidden 2
                                  tf.keras.layers.Dense(16,activation=tf.nn.relu), # Hidden 3
                                  tf.keras.layers.Dense(1)]) # Output
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
model_deep2.compile(optimizer=opt,loss='mse',metrics=['accuracy'])
hd2=model_deep2.fit(x_train2,y_train2,epochs=100,batch_size=16)
```

```
model_deep1.summary()
```

```
model_deep3=tf.keras.models.Sequential([#tf.keras.layers.Flatten(x),
                                  tf.keras.layers.Dense(128,activation=tf.nn.relu), # Hidden 1
                                  tf.keras.layers.Dropout(0.5),
                                  tf.keras.layers.Dense(16,activation=tf.nn.relu),# Hidden 2
                                  tf.keras.layers.Dense(16,activation=tf.nn.relu), # Hidden 3
                                  tf.keras.layers.Dense(1)]) # Output
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
model_deep3.compile(optimizer=opt,loss='mse',metrics=['accuracy'])
hd3=model_deep3.fit(x_train3,y_train3,epochs=100,batch_size=48)
```

# **Model Validation**

```
ypd1=model_deep1.predict(x_train1)
ypd2=model_deep2.predict(x_train2)
ypd3=model_deep3.predict(x_train3)
```
